```python
# read the csv files 
import pandas as pd
import sys, os, re, glob
import numpy as np

from scipy.interpolate import splev, splrep

import matplotlib.pyplot as plt
import matplotlib as mpl

#import seaborn as sns

import seaborn as sns

sns.set(style="ticks")
sns.set_context("poster")


smoothing=0.0
smoothing1=0.002
smoothing2=0.001
smoothing3=0.2



ymax =1.5
outdir='RTD_call'
inputfile='combineLogStats_cp.csv'
inputfile_obs='limits_comb3_cards_cp_Jul21_cv_1p0.csv'

CHANNEL= {
    "combineLogStats_cp.csv"   :"comb3",  #"2lss_mm_em_3l",
}

chan = CHANNEL[inputfile]

xsec_cp = pd.DataFrame.from_csv("xsec_cp.csv",sep=",", index_col=None) # read xsec from hbb thesis
limits_sm = pd.DataFrame.from_csv('%s'%inputfile,sep=",", index_col=None)
limits_obs = pd.DataFrame.from_csv('%s'%inputfile_obs,sep=",", index_col=None)

limits_sm['obs']=limits_obs['obs']
xsec = pd.DataFrame()

xsec_cp['ttH'] = xsec_cp['ttH']/0.582 # Remove H->bb BR from ttH

xsec=xsec_cp # These are production xs now; note the cp=1, cv=1 are SM xsec
xsec
```

```python
limits_sm
```

```python
#add up xsections and create new columns for (th, tth and total) xsec*BR


SM_BR_ww = 0.214
SM_BR_zz = 0.026
SM_BR_tt = 0.063

xsec['tHqww']= xsec['tHq']*SM_BR_ww
xsec['tHqzz']= xsec['tHq']*SM_BR_zz
xsec['tHqtt']= xsec['tHq']*SM_BR_tt

xsec['tHWww']= xsec['tHW']*SM_BR_ww
xsec['tHWzz']= xsec['tHW']*SM_BR_zz
xsec['tHWtt']= xsec['tHW']*SM_BR_tt

xsec['ttHww']= xsec['ttH']*SM_BR_ww
xsec['ttHzz']= xsec['ttH']*SM_BR_zz
xsec['ttHtt']= xsec['ttH']*SM_BR_tt

xsec['xsec_th'] = (xsec['tHq']+xsec['tHW'])*(SM_BR_ww+SM_BR_zz+SM_BR_tt)
xsec['xsec_tth'] = (xsec['ttH'])*(SM_BR_ww+SM_BR_zz+SM_BR_tt)
xsec['xsec_tot'] = (xsec['tHq']+xsec['tHW']+xsec['ttH'])*(SM_BR_ww+SM_BR_zz+SM_BR_tt)

xsec.sort_index(axis=1, inplace=True)

xsec.to_csv("xsecs_tH_ttH_cp.csv")

xsec
```

```python
#getting the limits on the cross section
#write the dataframe to a csv file

xsec_lim = pd.DataFrame()
xsec_lim['cp'] = xsec['cp']
xsec_lim['cv'] = xsec['cv']

for col in ['twosigdown','onesigdown', 'exp','onesigup','twosigup','obs']:
    xsec_lim['%s'%col]=limits_sm['%s'%col]*xsec['xsec_tot']
       
xsec_lim
```

```python
#read data and make extrapolations 

x = sorted(list(set(limits_sm.cp.values.tolist())))
x2 = np.linspace(-1,1,100)

spline_xsec_tot = splrep(x, xsec.loc[xsec.cv==1.0].xsec_tot, s=smoothing)
spline_xsec_th  = splrep(x, xsec.loc[xsec.cv==1.0].xsec_th, s=smoothing)
spline_xsec_tth = splrep(x, xsec.loc[xsec.cv==1.0].xsec_tth, s=smoothing)

y2_xsec_tot = splev(x2, spline_xsec_tot)
y2_xsec_tth = splev(x2, spline_xsec_tth)
y2_xsec_th  = splev(x2, spline_xsec_th)

spline_twosigdown = splrep(x,xsec_lim.twosigdown, s=smoothing1, k=3)
spline_onesigdown = splrep(x,xsec_lim.onesigdown, s=smoothing, k=3)
spline_exp        = splrep(x,xsec_lim.exp,        s=smoothing, k=3)
spline_onesigup   = splrep(x,xsec_lim.onesigup,   s=smoothing, k=3)
spline_twosigup   = splrep(x,xsec_lim.twosigup,   s=smoothing1, k=3)

y_twosigdown      = splev(x2, spline_twosigdown)
y_onesigdown      = splev(x2, spline_onesigdown)
y_exp             = splev(x2, spline_exp)
y_onesigup        = splev(x2, spline_onesigup)
y_twosigup        = splev(x2, spline_twosigup)

y_obs             = xsec_lim.obs
```

```python
fig, ax = plt.subplots(1)

# Configure axes                                                                                                                                                                                        
ax.get_xaxis().set_tick_params(which='both', direction='in')
#ax.get_xaxis().set_major_locator(mpl.ticker.MultipleLocator(1.0))
#ax.get_xaxis().set_minor_locator(mpl.ticker.MultipleLocator(0.25))
ax.set_xlim(-1.0, 1.0)
#ax.grid(False)



ax.set_ylim(0.,ymax)
ax.set_yscale("linear", nonposy='clip')
#ax.get_yaxis().set_tick_params(which='both', direction='in')
#ax.get_yaxis().set_major_locator(mpl.ticker.MultipleLocator(0.2))
#ax.get_yaxis().set_minor_locator(mpl.ticker.MultipleLocator(0.05))

ax.plot(x2, y_exp, "--", lw=2, color='black',zorder=4)

ax.fill_between(x2, y_onesigdown, y_onesigup,   facecolor='dodgerblue', edgecolor='k', alpha=0.8,zorder=0)
ax.fill_between(x2, y_twosigdown, y_onesigdown, facecolor='lightblue', edgecolor='k', alpha=0.8,zorder=0)
ax.fill_between(x2, y_twosigup,   y_onesigup,   facecolor='lightblue', edgecolor='k', alpha=0.8,zorder=0)

ax.plot(x, y_obs, ".",linestyle='-',lw=2, color='black',zorder=4)

ax.plot(x2, y2_xsec_tot, lw=1.5, label='dummy',color='black' ,linestyle='-.')

#if split_xsecs:                                                                                                                                                                                        
ax.plot(x2, y2_xsec_tth, lw=0.5, label='dummy',color='black', linestyle='dashed')
ax.plot(x2, y2_xsec_th,  lw=0.5, label='dummy',color='black', linestyle='dotted')

ax.set_xlabel('$ cos (\\alpha_{CP})$',fontsize=24,labelpad=20) # alpha is the CP-phase angle 
ax.set_ylabel('$\sigma\\times$ BR [pb]',fontsize=24,labelpad=20)
def print_text(x, y, text, fontsize=12, addbackground=True):
        if addbackground:
            ptext = plt.text(x, y, text, fontsize=fontsize, transform=ax.transAxes, backgroundcolor='white')
            ptext.set_bbox(dict(alpha=0.8, color='white'))
        else:
            ptext = plt.text(x, y, text, fontsize=fontsize, transform=ax.transAxes)
            
print_text(0.03, 1.02, "$\mathbf{CMS}$ Preliminary", 20, addbackground=False)
print_text(0.65, 1.02, "%.1f $\mathrm{fb}^{-1}$ (13 TeV)"% (35.9), 20, addbackground=False)

print_text(0.06, 0.92, "$\mathrm{pp}\\to\mathrm{tH}+\mathrm{t\\bar{t}H}$" , 15)
print_text(0.06, 0.86, ("$\mathrm{H}\\to\mathrm{W}\mathrm{W}/\mathrm{Z}\mathrm{Z}"
                            "/\mathrm{\\tau}\mathrm{\\tau}$"),15)
#print_text(0.06, 0.80, "%s"%chan)


# Cosmetics                                                                                                                                                                                             
import matplotlib.patches as mpatches

twosigpatch = mpatches.Patch(color='lightblue',     label='$\pm2$ standard dev.')
onesigpatch = mpatches.Patch(color='dodgerblue', label='$\pm1$ standard dev.')

obsline = mpl.lines.Line2D([], [], color='black', linestyle='-',
                               label='Observed limit ($\sigma\\times\mathrm{BR}$)', marker='.',
                               markersize=14, linewidth=2)

explabel = 'SM Higgs exp.limit ($\sigma\\times\mathrm{BR}$)'
expline = mpl.lines.Line2D([], [], color='black', linestyle='--', label=explabel, linewidth=2.0)

xsline = mpl.lines.Line2D([], [], color='black', lw=1.5, linestyle='-.',
                               label='$\sigma^{theo.}_{\mathrm{tH+t\\bar{t}H}}\\times \mathrm{BR}$')

tthxsline = mpl.lines.Line2D([], [], color='black', lw=1, linestyle='dashed',
                       label='$\sigma^{theo.}_{\mathrm{t\\bar{t}H}}\\times \mathrm{BR}$')
thxsline = mpl.lines.Line2D([], [], color='black', lw=1, linestyle='dotted',
                       label='$\sigma^{theo.}_{\mathrm{tH}}\\times \mathrm{BR}$')
            

# Legend                                                                                                                                                                                                
legentries = [obsline]                                                                                                                                                                            
legentries += [expline,onesigpatch,twosigpatch]
legentries += [xsline]                                                                                                                                                                            
legentries += [tthxsline, thxsline]
legend = plt.legend(handles=legentries, fontsize=14, frameon=True, loc='upper center', framealpha=0.8)#, bbox_to_anchor=(1,1))
legend.get_frame().set_facecolor('white')
legend.get_frame().set_linewidth(0)

#plt.plot(x,xsec.xsec_tot,'o',x2,y2_xsec_tot)
plt.show()
```

```python
# Save to pdf/png
outfile = os.path.join(outdir, os.path.splitext(os.path.basename(inputfile))[0])
plt.savefig("%s.pdf"%outfile, bbox_inches='tight')
plt.savefig("%s.png"%outfile, bbox_inches='tight', dpi=300)
print "...saved plots in %s.pdf/.png" % outfile
```
